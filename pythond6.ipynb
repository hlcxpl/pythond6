```python
import pandas as pd

# Cargar datos de cada hoja en DataFrames separados
excel_file = pd.ExcelFile("US_Regional_Sales_Data.xlsx")

sheet_names = excel_file.sheet_names

# dataframes = {}  # Diccionario para almacenar los DataFrames

# for sheet_name in sheet_names:
#     dataframes[sheet_name] = excel_file.parse(sheet_name)

sheet_names
```

```python
df_sales = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Sales Orders Sheet')

df_customers = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Customers Sheet')

df_stores = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Store Locations Sheet')

df_products = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Products Sheet')

df_teams = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Sales Team Sheet')
```

```python
df_base = df_sales.merge(df_customers, validate = 'many_to_one').\
                   merge(df_stores, validate = 'many_to_one').\
                   merge(df_products, validate = 'many_to_one').\
                   merge(df_teams, validate = 'many_to_one')
df_base
```

```python
# Calcular ProcurementDays, ShippingDays, DeliveryDays y CustomerDays
df_base["ProcurementDays"] = (df_base["OrderDate"] - df_base["ProcuredDate"]).dt.days
df_base["ShippingDays"] = (df_base["ShipDate"] - df_base["OrderDate"]).dt.days
df_base["DeliveryDays"] = (df_base["DeliveryDate"] - df_base["ShipDate"]).dt.days
df_base["CustomerDays"] = df_base["ShippingDays"] + df_base["DeliveryDays"]
df_base
```

```python
# Agregar la columna CustomerDaysInterval
df_base["CustomerDaysInterval"] = pd.cut(
    df_base["CustomerDays"],
    bins=[0, 15, 30, 45, 60, 75, 90],
    labels=[
        "0 to 15 days",
        "15 to 30 days",
        "30 to 45 days",
        "45 to 60 days",
        "60 to 75 days",
        "75 to 90 days",
    ],
)

df_base[["CustomerDays", "CustomerDaysInterval"]]
```

```python
# Crear una tabla pivote para el conteo de órdenes
pivot_table1 = pd.pivot_table(
    df_base,
    values="OrderNumber",
    index="Sales Team",
    columns="CustomerDaysInterval",
    aggfunc="count",
    fill_value=0,
    margins=True,
)
pivot_table1
```

```python
# Calcular GrossMargin
df_base["GrossMargin"] = df_base["Order Quantity"]*(df_base["Unit Price"]*(1-df_base["Discount Applied"])-df_base["Unit Cost"])
df_base
```

```python
# Definir los intervalos y etiquetas en formato de porcentaje
intervalos_margen = [0, 100, 1000, 10000, 100000]
etiquetas_margen = ['5%', '10%', '15%', '20%']

# Discretizar y asignar etiquetas en formato de porcentaje a la columna "CommissionsPercentage"
df_base["CommissionsPercentage"] = pd.cut(df_base["GrossMargin"], bins=intervalos_margen, labels=etiquetas_margen, right=False)

# Mostrar las columnas "GrossMargin" y "CommissionsPercentage"
df_base
```

```python
# Mantener la columna "CommissionsPercentage" en formato de porcentaje entero para su visualización
# y convertirla a valores numéricos enteros para cálculos
#df_base["CommissionsPercentageNumeric"] = df_base["CommissionsPercentage"].str.rstrip('%').astype(int)

# Calcular la columna "CommissionsAmount" como el producto de "GrossMargin" y "CommissionsPercentageNumeric"
df_base['CommissionsAmount'] = df_base['GrossMargin'] * df_base['CommissionsPercentageNumeric'] // 100

# mostrar "CommissionsPercentage" en formato de porcentaje 
df_base['CommissionsPercentage'] = df_base['CommissionsPercentageNumeric'].map(lambda x: f'{x}%')
df_base[['GrossMargin', 'CommissionsPercentage', 'CommissionsAmount']]

```

```python
# Calcular la columna "NetMargin" restando "CommissionsAmount" de "GrossMargin"
df_base['NetMargin'] = df_base['GrossMargin'] - df_base['CommissionsAmount']

# Mostrar las columnas "GrossMargin", "CommissionsAmount" y "NetMargin"
df_base[['GrossMargin', 'CommissionsAmount', 'NetMargin']]
```
