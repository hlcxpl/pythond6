1. Carga los datos de cada pestaña del archivo US_Regional_Sales_Data.xlsx en un
DataFrame independiente. (hint: Puedes obtener los nombres de las hojas usando
pd.ExcelFile(archivo).sheet_names)


```python
import pandas as pd

# Cargar datos de cada hoja en DataFrames separados
excel_file = pd.ExcelFile("US_Regional_Sales_Data.xlsx")

sheet_names = excel_file.sheet_names

# dataframes = {}  # Diccionario para almacenar los DataFrames

# for sheet_name in sheet_names:
#     dataframes[sheet_name] = excel_file.parse(sheet_name)

sheet_names
```

```python
df_sales = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Sales Orders Sheet')

df_customers = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Customers Sheet')

df_stores = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Store Locations Sheet')

df_products = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Products Sheet')

df_teams = pd.read_excel("US_Regional_Sales_Data.xlsx", sheet_name = 'Sales Team Sheet')
```

2. Cruza todos los DataFrames usando validación many_to_one, y guarda el resultado
en una nueva variable llamada df_base

```python
df_base = df_sales.merge(df_customers, validate = 'many_to_one').\
                   merge(df_stores, validate = 'many_to_one').\
                   merge(df_products, validate = 'many_to_one').\
                   merge(df_teams, validate = 'many_to_one')
df_base
```

3. Agrega las siguientes columnas a df_ base:
Columna Cálculo
ProcurementDays OrderDate-ProcuredDate
ShippingDays ShipDate-OrderDate
DeliveryDays DeliveryDate-ShipDate
CustomerDays ShippingDays+DeliveryDays

```python
# Calcular ProcurementDays, ShippingDays, DeliveryDays y CustomerDays
df_base["ProcurementDays"] = (df_base["OrderDate"] - df_base["ProcuredDate"]).dt.days
df_base["ShippingDays"] = (df_base["ShipDate"] - df_base["OrderDate"]).dt.days
df_base["DeliveryDays"] = (df_base["DeliveryDate"] - df_base["ShipDate"]).dt.days
df_base["CustomerDays"] = df_base["ShippingDays"] + df_base["DeliveryDays"]
df_base
```

4. Agrega una nueva columna discreta al df_base, con el nombre CustomerDaysInterval
que clasifique los valores de la columna CustomerDays en los siguientes intervalos:
○ 0 to 15 days
○ 15 to 30 days
○ 30 to 45 days
○ 45 to 60 days
○ 60 to 75 days
○ 75 to 90 days
hint: Utiliza pd.cut con estas etiquetas de Intervalo. Puedes convertir previamente
la columna CustomerDays a valores numéricos con pd.to_numeric(). Considera que
en este caso, el resultado es entregado en formato 10**9 segundos, por lo que
debes dividir el resultado por este valor

```python
# Agregar la columna CustomerDaysInterval
df_base["CustomerDaysInterval"] = pd.cut(
    df_base["CustomerDays"],
    bins=[0, 15, 30, 45, 60, 75, 90],
    labels=[
        "0 to 15 days",
        "15 to 30 days",
        "30 to 45 days",
        "45 to 60 days",
        "60 to 75 days",
        "75 to 90 days",
    ],
)

df_base[["CustomerDays", "CustomerDaysInterval"]]
```

5. Utiliza una tabla pivote para mostrar el conteo de órdenes agrupando por Sales
Team en las filas y CustomerDaysInterval en las columnas. Este reporte debe
cuadrar con los datos de la pestaña Reporte1 del archivo reportes.xlsx

```python
# Crear una tabla pivote para el conteo de órdenes
pivot_table1 = pd.pivot_table(
    df_base,
    values="OrderNumber",
    index="Sales Team",
    columns="CustomerDaysInterval",
    aggfunc="count",
    fill_value=0,
    margins=True,
)
pivot_table1
```

6. Agrega la siguiente columna calculada a la base:
KPI Cálculo
GrossMargin OrderQuantity*(UnitPrice*(1-Discount Applied)-UnitCost)
Visión de Negocio: El margen bruto es un indicador financiero que muestra la diferencia entre
los ingresos generados por la venta de un producto o servicio y los costos directos asociados
con su producción o adquisición.


```python
# Calcular GrossMargin
df_base["GrossMargin"] = df_base["Order Quantity"]*(df_base["Unit Price"]*(1-df_base["Discount Applied"])-df_base["Unit Cost"])
df_base["GrossMargin"]
```

7. En esta parte debes aplicar un porcentaje a la columna GrossMargin, pero ese
porcentaje dependerá del intervalo en que se encuentre GrossMargin.
KPI Cálculo
CommissionsPercentage % de comisión según la tabla de comisiones
hint: Utiliza pd.cut con labels numéricas para que la columna resultante sea de tipo float.


```python
# Definir los intervalos y etiquetas en formato de porcentaje
intervalos_margen = [0, 100, 1000, 10000, 100000]
etiquetas_margen = ['5%', '10%', '15%', '20%']

# Discretizar y asignar etiquetas en formato de porcentaje a la columna "CommissionsPercentage"
df_base["CommissionsPercentage"] = pd.cut(df_base["GrossMargin"], bins=intervalos_margen, labels=etiquetas_margen, right=False)

# Mantener la columna "CommissionsPercentage" en formato de porcentaje entero para su visualización
# y convertirla a valores numéricos enteros para cálculos
df_base["CommissionsPercentageNumeric"] = df_base["CommissionsPercentage"].str.rstrip('%').astype(int)

# Mostrar las columnas "GrossMargin" y "CommissionsPercentage"
df_base["CommissionsPercentage"]
```

8. Calcula el monto de la comisión, utilizando la Tabla del Anexo.
KPI Cálculo
CommissionsAmount GrossMargin*CommissionsPercentage

```python


# Calcular la columna "CommissionsAmount" como el producto de "GrossMargin" y "CommissionsPercentageNumeric"
df_base['CommissionsAmount'] = df_base['GrossMargin'] * df_base['CommissionsPercentageNumeric'] // 100

# mostrar "CommissionsPercentage" en formato de porcentaje solo modo estetico
df_base['CommissionsPercentage'] = df_base['CommissionsPercentageNumeric'].map(lambda x: f'{x}%')
df_base[['GrossMargin', 'CommissionsPercentage', 'CommissionsAmount']]

```

9. Calcula la comisión sobre el margen bruto.
KPI Cálculo
NetMargin GrossMargin-CommissionsAmount

```python
# Calcular la columna "NetMargin" restando "CommissionsAmount" de "GrossMargin"
df_base['NetMargin'] = df_base['GrossMargin'] - df_base['CommissionsAmount']

# Mostrar las columnas "GrossMargin", "CommissionsAmount" y "NetMargin"
df_base[['GrossMargin', 'CommissionsAmount', 'NetMargin']]
```

10. Utilizando una tabla pivote, muestra la suma de los valores GrossMargin, NetMargin y
CommissionAmount agrupando por Sales Team en las filas. Este reporte debe cuadrar con
los datos de la pestaña Reporte2 del archivo reportes.xlsx

```python
# Crear una tabla pivote para sumar los valores GrossMargin, NetMargin y CommissionAmount por Sales Team
pivot_table = pd.pivot_table(df_base, 
                             values=['GrossMargin', 'NetMargin', 'CommissionsAmount'], 
                             index='Sales Team', 
                             aggfunc='sum')

# Mostrar la tabla pivote
pivot_table
```
